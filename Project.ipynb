```python
import pandas as pd
#Importing the dataset
data=pd.read_csv("IEA GEVO 2023 data.csv")
#Head of the dataset
print("The head of the dataset is:\n",data.head(5),'\n')
#Instances
print("There are ",len(data.axes[0])," instances")
```

```python
#Variables
print("There are ",len(data.axes[1])," variables. \n")

#Column names
print("The columns of the dataset are\n",data.columns)
```

```python
#Information
print("The information about the data\n",data.info())
```

```python
#Description
print("The description of the dataset\n",data.describe())
```

```python
#Checking for null values
data.isnull().any()
```

```python
#Checking for distinct values in each column
print("Distinct values in region:\n",data['region'].unique(),"\n")
print("Distinct values in category:\n",data['category'].unique(),"\n")
print("Distinct values in parameter:\n",data['parameter'].unique(),"\n")
print("Distinct values in mode:\n",data['mode'].unique(),"\n")
print("Distinct values in powertrain:\n",data['powertrain'].unique(),"\n")
print("Distinct values in unit:\n",data['unit'].unique(),"\n")
print("Distinct values in year:\n",data['year'].unique(),"\n")
```

```python
#Deleting year which are not applicable for analysis
year_to_delete=[2010,2011,2012,2025,2030]
df=data.loc[~data['year'].isin(year_to_delete)]

#Deleting regions which are not applicable for analysis
region_to_delete=['EU27','Other Europe','Rest of the world','World']
df=df.loc[~df['region'].isin(region_to_delete)]

#Deleting parameters which are not applicable for analysis
parameter_to_delete=['EV sales share', 'EV stock share','EV stock','Oil displacement, million lge']
df=df.loc[~df['parameter'].isin(parameter_to_delete)]

#Deleting categorys which are not applicable for analysis
category_to_delete=['Projection-STEPS','Projection-APS']
df=df.loc[~df['category'].isin(category_to_delete)]
print(df)
#df.to_csv("C:/Users/pc7/Desktop/Belinda/EVdata.csv")
```

```python
#univariate analysis
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['parameter'])
plt.title('Distribution of parameter')
plt.xticks(rotation='vertical')
```

```python
#univariate analysis
sns.countplot(x=df['mode'])
```

```python
#univariate analysis
x = df['year'].value_counts()
plt.pie(x.values,
        labels=x.index,
        autopct='%1.1f%%')
plt.legend(title='Years', labels=x.index, loc='center left',bbox_to_anchor=(1.2, 0.5),ncol=1)
plt.show()
```

```python
#Bivariate
plt.figure(figsize=(15, 5))
sns.barplot(x=df['mode'], y=df['value'])
```

```python
#Bivariate
sns.countplot(x=df['year'],hue=df['mode'])
```

```python
#multivariate
numeric_columns=df.select_dtypes(include=['float64','int64'])
correlation_matrix=numeric_columns.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Regionwise Entry of records
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.scatter(df["region"],df["year"])
# adjusting name,size and location of title
plt.title("Region vs Year",size=25,y=1.05,x=0.45)

# adjusting name and size of x and y labels
plt.xlabel("Regions",size=20);plt.ylabel("Year",size=20)

# adjusting name and size of x and y tick labels
plt.xticks(rotation=90);plt.yticks(size=15)

# grid on
plt.grid()
```

```python
#Regionwise Entry of records
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.scatter(df["region"],df["parameter"])
# adjusting name,size and location of title
plt.title("Region vs Parameter",size=25,y=1.05,x=0.45)

# adjusting name and size of x and y labels
plt.xlabel("Regions",size=20);plt.ylabel("Parameters",size=20)

# adjusting name and size of x and y tick labels
plt.xticks(rotation=90);plt.yticks(size=15)

# grid on
plt.grid()
```
